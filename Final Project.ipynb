```python
## import modules
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
penguins = pd.read_csv('palmer_penguins.csv')
penguins.head()
```

```python
# want to reduce penguin species to Adelie, Gentoo, and Chinstrap
penguins['Species'] = penguins['Species'].apply(lambda x: x.split()[0])
penguins
```

```python
from sklearn.model_selection import train_test_split
```

```python
def split_data(df, X_cols, y_col, test_size):
    # X_cols must be a list of strings
    # y_col must be a string
    X = df[X_cols]
    y = df[y_col]    
    
    # return X_train, X_test, y_train, y_test
    return train_test_split(X, y, test_size=test_size)
```

```python
X_cols = ['Island', 'Culmen Length (mm)', 'Culmen Depth (mm)', 'Flipper Length (mm)', 'Body Mass (g)', 
         'Sex']
y_col = 'Species'
X_train, X_test, y_train, y_test = split_data(penguins, X_cols, y_col, 0.20)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# cleaning data for y -> encoding data
def encode_data(arr):
    le = LabelEncoder()
    le.fit(arr)
    return le.classes_, le.transform(arr)
```

```python
y_train_classes, y_train = encode_data(y_train)
y_test_classes, y_test = encode_data(y_test)
```

```python
 def encode_columns(df, cols): # encode qualitative columns in df
        # cols must be list of strings
        # cols must contain valid cols
        for col in cols:
            le = LabelEncoder()
            df[col] = le.fit_transform(df[col])
        return df
```

```python
X_train = encode_columns(X_train, ['Island', 'Sex'])
X_test = encode_columns(X_train, ['Island', 'Sex'])
```
